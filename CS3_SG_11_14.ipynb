#  Hospital Billing System

Apply encapsulation, inheritance, and method overriding in Python to create a hospital billing system that calculates different discounts based on patient type.

**Situation:** The hospital serves different types of patients:
- Regular Patients pay the full amount.   
- Senior Citizens receive a 20% discount.
- PhilHealth Members receive 15% discount.


### create a Python program with:

1. class Patient with the following features  
               **Attributes:** name, bill_amount  
               **Methods:**   Getter and setter methods  and
                                 calculate_bill() (returns full bill)
            
2.  child classes:
 - SeniorPatient
 - PhilHealthPatient

Overrides the calculate_bill() method in each subclass to apply the correct discount.



```python
class Patient:
  def __init__(self, name, bill):
    self._name = name
    self._bill_amount = 0
    self.bill_amount =bill

  @property
  def name(self):
    return self._name

  def calculate_bill(self):
    return self._bill_amount

  @property
  def bill_amount(self):
    return self._bill_amount

  @bill_amount.setter
  def bill_amount(self, bill_amount):
    if not bill_amount >= 0:
      print('Bill amount cannot be negative. Setting to 0')
      bill_amount = 0
      return

    self._bill_amount = bill_amount


  def __repr__(self):
    return (f'{self.__class__.__name__} {self.name} has a final '
      f'bill of Php{self.calculate_bill():.2f}')


class SeniorPatient(Patient):
  def calculate_bill(self):
    return super().calculate_bill() * 0.8

class PhilHealthPatient(Patient):
  def calculate_bill(self):
    return super().calculate_bill() * 0.85
```

Let's try it:

```python


jc = Patient("Juan dela Cruz", 15000)
ms = SeniorPatient("Maria Santos", 15000)
ped = PhilHealthPatient("Pedro Reyes", 15000)

for patient in [jc, ms, ped]:
  print(patient)
```

# Abstraction

In programming, Abstraction allows us to create "blueprints" that force other classes to follow certain rules without worrying about the messy details inside.

To do abstraction in Python, we use a special tool called `ABC` (Abstract Base Class). An abstract class is a class that cannot be instantiated (you can't create an object directly from it). It exists only to be a blueprint.

The `@abstractmethod` looks like a weird "comment" or a glitch in the code. In Python, this is called a Decorator.
Think of it as a "Sticky Note" that the lead programmer leaves for everyone else.

 What is `@abstractmethod`?
It is a Strict Rule or a Contract. When you put this "sticky note" above a method, you are telling Python:

> "Every child class that inherits from this MUST write their own version of this action. If they don't, I won't let the program run."


## Now  let's make make the character classes with abstraction

```python
from abc import ABC, abstractmethod

class Hero(ABC):
  @abstractmethod
  def special_move(self):
    pass


class Warrior(Hero):
  def special_move(self):
    print('Warrior swings a mighty sword!')

class Mage(Hero):
  def special_move(self):
    print('Mage casts a powerful spell!')

class Archer(Hero):
  def special_move(self):
    print('Archer shoots a precise arrow!')


def perform_move(character):
  character.special_move()

characters =[Warrior(), Mage(), Archer()]

for c in characters:
  perform_move(c)
```

Let's make ABC for the vehicles


```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

  @abstractmethod
  def move(self):
    pass

class Car(Vehicle):
  def move(self):
    print('The care drives on the road.')

class Boat(Vehicle):
  def move(self):
    print('The boat sails on water.')

class Drone(Vehicle):
  def move(self):
    print('The drone flies in the sky.')


def start_trip(vehicle):
  vehicle.move()

for vehicle in [Car(), Boat(), Drone()]:
  start_trip(vehicle)
```

## Let's make the potions!

Create the Python code that:

1. Has three classes: **HealthPotion**, **ManaPotion**, **SpeedBoost**.
Each class should have a method called `use()`:

    - HealthPotion’s use() method  prints "You regain 50 HP!"

    - ManaPotion’s use() method prints "You restore 30 MP!"

    - SpeedBoost’s use() method  prints "Your speed increases for 10 seconds!"

2. Has a function apply_item(item), with the parameter item,  that calls the method use() :  item.user();

3. Creates one object of each class.

4. Stores the objects in a list, potions,  and calls apply_item() for each object using a loop.

5. Use an *Abstract Base Class* for the potions


```python
from abc import ABC, abstractmethod

class Potion:
  @abstractmethod
  def use(self):
    pass

class HealthPotion(Potion):
  def use(self):
    print("You regain 50 HP.")

class ManaPotion(Potion):
  def use(self):
    print("You restore 30 MP.")

class SpeedBoost(Potion):
  def use(self):
    print('You get an extra turn')

def apply_item(item):
  item.use()

for item in [HealthPotion(), ManaPotion(), SpeedBoost()]:
  apply_item(item)
```
