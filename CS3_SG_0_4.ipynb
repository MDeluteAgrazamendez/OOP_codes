## Lesson 1 - Calculator: from functions to class

```python
def add(a,b):
  return a+ b

add(3,5)
```

```python
def mult(a,b):
  return a*b

mult(3,4)
```

```python
class Calc:
  def __init__(self, num1=0, num2=0):
    self.x = num1
    self.y = num2

  def add(self):
    return self.x + self.y

  def mult(self):
    return self.x * self.y


c = Calc(2,3)
print(c.add())
print(c.mult())
print(Calc(2,3).add())
```

```python
class Calc2(Calc):
  def setx(self, num):
    self.x = float(num)
  def sety(self, num):
    self.y = float(num)

c = Calc2()
# c.setx(input("x: "))
# c.sety(input("y: "))
print(c.add())
print(c.mult())
```

```python
class Calc3(Calc2):
  def add(self, a=None, b=None):
    if a is None:
      a = self.x
    if b is None:
      b = self.y

    return a +b

  def mult(self, a=None, b=None):
    if a is None:
      a = self.x
    if b is None:
      b = self.y
    return a*b

c = Calc3()
print(c.add(1,2))
print(c.mult(4,5))
c = Calc3(3,2)
print(c.add())
print(c.mult())
```

sum of all squared numbers from 1 to n.

```python
def sum_of_squared(n):
  if n<1:
    return "N/A"
  else:
    sum = 0
    for i in range(n):
      sum += i**2
    return sum

n=3
print(f"the sum of the first {n} squared numbers is {sum_of_squared(n)}")
```
